```python
#exploring data relations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import scipy.optimize

#read in the CSV data using Pandas
data = pd.read_csv("C:/Users/nicol/Uni 2019-2020/Python 2020/G2.Galaxy.Data.csv", delimiter = ',')
```

```python
#Plots 2: stellar mass per redshift
plt.scatter(data['stellarMass'], data['redshift'], marker='.', color='firebrick')
plt.xlabel('Stellar Mass')
plt.ylabel("Redshift")
plt.title('Stellar Mass per Redshift')
plt.savefig("CTA200.stellarmassperz.pdf")
plt.show()

#Try log 
plt.scatter(data['stellarMass'], data['redshift'], marker='.', color='firebrick')
plt.yscale('log')
plt.xlabel('Stellar Mass')
plt.ylabel("Redshift")
plt.title('Stellar Mass per Redshift: Log Axis')
plt.savefig("CTA200.stellarmassperz.pdf")
plt.show()
```

```python
#filter the redshift for in situ 
#Check if the sfr increases 
#According to the Oser 2 phases paper 
redshift = data['redshift']
redshift_insitu = []
index = []
for i in range(len(redshift)):
    if redshift[i] > 2 and redshift[i] < 3: 
        redshift_insitu.append(redshift[i])
        index.append(i)
```

```python
print(redshift_insitu)
print(index)
```

```python
sfr = data['sfr']
sfr_insitu = []

for index in index: 
    sfr_insitu.append(sfr[index])
```

```python
#plot 
plt.scatter(sfr_insitu, redshift_insitu, marker='.', color='firebrick')
plt.xlabel('Star Formation Rate')
plt.ylabel("Redshift")
plt.title('Star Formation Rate per Redshift: In Situ')
plt.show()
```

```python
#Checking z = 4 
redshift = data['redshift']
redshift_insitu = []
index = []
for i in range(len(redshift)):
    if redshift[i] > 3 and redshift[i] <= 4: 
        redshift_insitu.append(redshift[i])
        index.append(i)
sfr = data['sfr']
sfr_insitu = []

for index in index: 
    sfr_insitu.append(sfr[index])
    #plot 
plt.scatter(sfr_insitu, redshift_insitu, marker='.', color='firebrick')
plt.xlabel('Star Formation Rate')
plt.ylabel("Redshift")
plt.title('Star Formation Rate per Redshift: Z ~ 4')
plt.show()
```

```python
#Habitability of Galaxies
#Star Formation Rate Studies 

#Plot: star formation rate per redshift
plt.scatter(data['redshift'], data['sfr'], marker='.', color='mediumvioletred')
plt.xlabel('z')
plt.ylabel("sfr")
plt.title('Star Formation Rate per Redshift')
plt.show()

#Narrow Redshift 
plt.scatter(data['redshift'], data['sfr'], marker='.', color='mediumvioletred')
plt.axis([3,5,0,150])
plt.xlabel('z')
plt.ylabel("sfr")
plt.title('Star Formation Rate per Redshift: Narrowed')
plt.show()

#try curve fit 
scipy.optimize.curve_fit(data['redshift'], data['sfr'])
```
